```python
away_gi_query = f'''
SELECT gi.*
FROM game_info gi
WHERE (gi.away_team = (SELECT away_team FROM game_info WHERE game_id = 222304091230)
        AND gi.date < '{date}')
   OR (gi.home_team = (SELECT away_team FROM game_info WHERE game_id = 222304091230)
        AND gi.date < '{date}')
ORDER BY gi.date DESC
LIMIT 10;
'''
away_gi_query_df = pd.read_sql(away_gi_query, con=conn)
away_gi_query_df
```

```python
away_gid_query = f'''
SELECT gi.game_id
FROM game_info gi
WHERE (gi.away_team = (SELECT away_team FROM game_info WHERE game_id = 222304091230)
        AND gi.date < '{date}')
   OR (gi.home_team = (SELECT away_team FROM game_info WHERE game_id = 222304091230)
        AND gi.date < '{date}')
ORDER BY gi.date DESC
LIMIT 10;
'''
away_gid_query_df = pd.read_sql(away_gid_query, con=conn)
away_gid_query_df


# F4_query = '''
# SELECT game_id, FGA, FTA, FG, '3P', TOV
# FROM team_stats
# '''

# away_F4_query = f'''
# SELECT *
# FROM {away_gid_query}
# JOIN {F4_query} 
#   ON {away_gid_query}.game_id = {F4_query}.game_id

# away_F4_query = f'''
# {away_gid_query}
# '''
# qu_test_df = pd.read_sql(away_F4_query, con=conn)
# qu_test_df

# att_qu = '''
# SELECT gi.game_id
# FROM game_info gi
# JOIN team_stats ON gi.game_id = team_stats.game_id
# WHERE (gi.away_team = (SELECT away_team FROM game_info WHERE game_id = 222304091230)
#         AND gi.game_id < 222304091230)
#    OR (gi.home_team = (SELECT away_team FROM game_info WHERE game_id = 222304091230)
#         AND gi.game_id < 222304091230)
# ORDER BY gi.date DESC
# LIMIT 10;
# '''
# att_qu_df = pd.read_sql(att_qu, con=conn)
# att_qu_df
```
