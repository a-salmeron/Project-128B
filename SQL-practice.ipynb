```python
query = '''
SELECT ps.*
FROM player_stats ps
JOIN game_info gi ON ps.game_id = gi.game_id
WHERE (gi.away_team = (SELECT away_team FROM game_info WHERE game_id = 222304091230)
  AND gi.date < (SELECT date FROM game_info WHERE game_id = 222304091230))
  OR (gi.home_team = (SELECT away_team FROM game_info WHERE game_id = 222304091230)
  AND gi.date < (SELECT date FROM game_info WHERE game_id = 222304091230))
ORDER BY gi.date DESC;
'''
query_df = pd.read_sql(query, con=conn)
query_df
```
